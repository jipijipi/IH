```python
import pandas as pd
```

```python
data = pd.read_csv('data/Melbourne_housing_FULL.csv')
data.isna().sum()
data
```

```python
data = data.dropna(subset=['Price'])
data.loc[:, 'Price'] = data['Price'].astype(int)

data
```

```python
type_counts = data['Type'].value_counts(normalize=True) * 100
type_counts
```

```python
import matplotlib.pyplot as plt

# Group by the number of rooms and calculate the median price
median_price_by_rooms = data.groupby('Rooms')['Price'].median()

# Plot the bar chart
median_price_by_rooms.plot(kind='bar', color='skyblue')
plt.title('Median Property Price by Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Median Price')
plt.show()
```
