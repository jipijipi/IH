```python


objects = {'P':(18,4),'U':(7,4),'E':(0,0),'C':(19,1),'V':(19,9)}

layout = [['room a', None, None],
          ['room b', 'room c', 'room d']]

def create_room(is_empty = False):
    room = []
    ceiling_tile = '-'
    wall_tile = '|'
    width = 20
    height = width // 2

    for i in range(height):

        if not is_empty:
            if i == 0 or i == height - 1:
                room.append(ceiling_tile * width)
            else:
                room.append(wall_tile + ' ' * (width - 2) + wall_tile)
        else:
            room.append(' ' * width)

    return room


def insert_objects(room, objects):
    for symbol, (x, y) in objects.items():
        room[y] = room[y][:x] + symbol + room[y][x + 1:]
    return room



#here 

def create_map(layout):
    map = []
    for row in layout:
        row_of_rooms = []
        for room in row:
            if room is None:
                row_of_rooms.append(create_room(True))
            else:
                row_of_rooms.append(insert_objects(create_room(), objects))
        map.append(row_of_rooms)
    return map


def display_map(map):
    final_map = ''

    for row_of_rooms in map:
        display_row = ''
        row_size = len(row_of_rooms[0])

        for i in range(row_size):
            for room in row_of_rooms:
                display_row += room[i]
            display_row += '\n'
        final_map += display_row

    print(final_map)

        

       


test_map = create_map(layout)
display_map(test_map)
```
