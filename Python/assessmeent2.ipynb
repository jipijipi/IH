```python
import pandas as pd

def analyze(factbook_pop: str, factbook_obesity: str) -> pd.DataFrame:
    pop_df = pd.read_csv(factbook_pop)
    obe_df = pd.read_csv(factbook_obesity)
    
    
    pop_df=pop_df.rename(columns={'Name':'Country', 'Value':'Population'})
    obe_df=obe_df.rename(columns={'Name':'Country', 'Value':'Obesity Rate'})
    
    pop_df=pop_df.drop(columns=['Pos'])
    obe_df=obe_df.drop(columns=['Pos'])
    
    
    df = pd.merge(pop_df, obe_df, on='Country')
    
    
    segment = df[(df['Obesity Rate'] > 20) & (df['Population'] > 10000000)].sort_values(by='Obesity Rate', ascending=False)
    
    top_10 = segment.head(10).reset_index(drop=True)
    top_10.index = top_10.index + 1
    
    return top_10

#analyze('data/c2119.csv','data/c2228.csv')
    
```

```python

df = pd.read_csv('data/attention.csv')


def how_many_times_better_is_focused(df: pd.DataFrame) -> float:
    """ Returns how many times better it is to focus than to divide one's attention, based on the average scores (returns the resulting ratio).
    """

    pivot = df.pivot_table(index='attention', values='score', aggfunc='mean')

    n_times = pivot.loc['focused', 'score'] / pivot.loc['divided', 'score']
    return n_times


def highest_average_solutions(df: pd.DataFrame) -> int:
    """ Returns the solutions id with the highest average score """

    pivot = df.pivot_table(index='solutions', values='score', aggfunc='mean')
    solution_id = pivot.idxmax().iloc[0]
    return solution_id


def top_10_scoring_subjects(df: pd.DataFrame) -> pd.DataFrame:

    df = df.pivot_table(index='subject', values='score', aggfunc='mean').sort_values(by=['score', 'subject'], ascending=[False,True]).reset_index().head(10)

    return df


top_10_scoring_subjects(df)
```
